##Análise mensal de compras

###Reduzindo o DataFrame orginal para apenas as colunas necessárias
As colunas de interesse são  a data da criação do pedido, valor pago e somente são consideradas as compras que não foram canceladas.

```python
df['status'].unique()
```

```python
df_trim = df[['date_created', 'status', 'paid_amount']]
df_trim
```

Apenas o status pago é considerado.

```python
df_trim_2 = df_trim.query("status == 'paid'")
df_trim_2
```

Criação de uma nova coluna apenas com o valor do mês da compra

```python
df_trim_2['month'] = df_trim_2['date_created'].apply(lambda x: x[5:7])
df_trim_2
```

Observe que há 12 valores únicos para o mês.

```python
df_trim_2['month'].unique()
```

Remoção da coluna data.

```python
df_trim_3 = df_trim_2[['month', 'paid_amount']]
df_trim_3
```

A dados foram rearranjados em ordem crescente do mês.

```python
df_trim_3 = df_trim_3.sort_values(by=['month'])
df_trim_3
```

Os valores de cada mês foram somados.

```python
df_trim_3 = df_trim_3.groupby(by='month').sum()
df_trim_3
```

```python
df_trim_3.describe()
```

```python
px.box(df_trim_3, y='paid_amount', labels={'paid_amount': 'Total mensal em R$'})
```

###Geração do gráfico de barras

```python
px.bar(df_trim_3,x=months, y='paid_amount', labels={'x':'Meses', 'paid_amount': 'Total mensal em R$'})
```

Conclusão: Fevereiro é o mês com menor receita, R\$ 74 mil, e Junho e Julho são os meses de maior receita, R\$ 157 mil.

##Pico de horário de vendas

```python
tabela = pd.read_csv('orders_combined_with_payments.csv', encoding='cp860')
```

```python
tabela_atualizada = tabela.loc[:, ['id','date_closed', 'date_last_updated', 'tags']]
tabela_atualizada['date_closed'] = tabela_atualizada['date_closed'].apply(lambda x: int(x[11:13]))
tabela_atualizada['date_last_updated'] = tabela_atualizada['date_last_updated'].apply(lambda x: int(x[11:13]))
tabela_atualizada
```

```python
#Calculando a média de horário
media = tabela_atualizada['date_closed'].mean()
print(media)
```

```python
tabela_atualizada['date_closed'] = tabela_atualizada['date_closed'].astype(int)
tabela_atualizada['date_closed'].plot(kind='hist', bins=24, title='Frequência de Vendas x Horário', edgecolor='black')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xlabel('Horário')
plt.ylabel('Vendas')
bins = range(24)
label = ['']*24
for i in range(24):
  label[i] = str(bins[i]) + 'h'
plt.xticks(bins, label)

plt.show()
```

```python
bins = [0, 6, 12, 18, 23]

tabela_atualizada['date_closed'] = tabela_atualizada['date_closed'].astype(int)
tabela_atualizada['date_closed'].plot(kind='hist', bins=bins, title='Frequência de Vendas x Período do Dia', edgecolor='black')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel('Período do Dia')
plt.ylabel('Vendas')

# Set x-tick labels for clarity
plt.xticks(bins, ['Madrugada', 'Manhã', 'Tarde', 'Noite', ''])

plt.show()
```

##Meios de pagamentos e parcelas

```python
df_test = pd.read_csv('orders_combined_with_payments.csv', encoding='cp860')
```

```python
df_meios = df_test[['payments_payment_method_id', 'payments_installment_amount', 'payments_installments']]
df_meios
#payments_payment_method_id: método de pagamento
#payments_installment_amount: valor da parcela(nos NaN, foi so uma parcela)
#payments_installments: quantidade de parcelas
```

```python
metodos_pagamento_counts = df_meios['payments_payment_method_id'].value_counts()
metodos_pagamento_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Método de Pagamento')
plt.ylabel('Transações')
#plt.title('Distribuição de Métodos de Pagamento')
plt.show()
```

```python
mean_installments_per_method = df_meios.groupby('payments_payment_method_id')['payments_installments'].mean()

plt.figure(figsize=(20, 12))
sns.barplot(x=mean_installments_per_method.index, y=mean_installments_per_method.values)
plt.xlabel('Método de Pagamento')
plt.ylabel('Número Médio de Parcelas')
#plt.title('Número Médio de Parcelas por Método de Pagamento')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='payments_installments', y='payments_installment_amount', data=df_meios)
plt.xlabel('Número de Parcelas')
plt.ylabel('Valor da Parcela')
#plt.title('Correlação entre Valor da Parcela e Número de Parcelas')
plt.show()
```

```python
df_meios['total_volume'] = df_meios['payments_installment_amount'] * df_meios['payments_installments']
volume_por_metodo = df_meios.groupby('payments_payment_method_id')['total_volume'].sum()

plt.figure(figsize=(8, 8))
plt.pie(volume_por_metodo, labels=volume_por_metodo.index, autopct='%1.1f%%', startangle=140)
#plt.title('Volume de Operações por Método de Pagamento')
plt.show()
```

```python
df_meios['volume_vendas_reais'] = df_meios['payments_installment_amount'] * df_meios['payments_installments']

sns.scatterplot(x='payments_installments', y='volume_vendas_reais', data=df_meios)

#plt.title('Volume de Vendas em Relação ao Número de Parcelas')
plt.xlabel('Número de Parcelas')
plt.ylabel('Volume de Vendas (R$)')

plt.show()
```
